```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.

import matplotlib.pyplot as plt
import seaborn as sns  # Required to apply Seaborn themes

# Apply a valid Seaborn or Matplotlib style
sns.set_theme(style="darkgrid")  # or try "whitegrid", "ticks", "dark", etc.
plt.rcParams['font.family'] = 'monospace'

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 11]

# Create the line plot
plt.plot(x, y, linestyle='--', marker='o', color='crimson', linewidth=2, markersize=8, label='Sample Line')

# Add title and labels
plt.title('Styled Line Plot', fontsize=14, fontweight='bold')
plt.xlabel('X-Axis', fontsize=12)
plt.ylabel('Y-Axis', fontsize=12)

# Add grid and legend
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()

```
