```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.

import matplotlib.pyplot as plt

# Product names and their corresponding sales in thousands
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Set custom font and style
plt.rcParams['font.family'] = 'serif'
plt.style.use('ggplot')  

# Create the bar plot
plt.bar(products, sales, color='teal')

# Add grid, labels, and title
plt.grid(True, which='major', axis='y', linestyle='--', alpha=0.7)
plt.title('Product Sales (in Thousands)', fontsize=14, fontweight='bold')
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
