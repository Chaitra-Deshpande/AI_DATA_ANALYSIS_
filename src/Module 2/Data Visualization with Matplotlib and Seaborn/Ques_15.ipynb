```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Select only numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr_matrix = numeric_tips.corr()

# Create the heatmap with advanced styling
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,                # The correlation matrix
    annot=True,                 # Display values on the heatmap
    cmap='coolwarm',            # Color map
    center=0,                   # Center the colormap at 0
    linewidths=0.5,             # Line thickness between cells
    linecolor='white',          # Color of the lines separating cells
    square=True,                # Make the heatmap square
    cbar_kws={'shrink': 0.8},   # Adjust the color bar size
    annot_kws={'size': 10, 'weight': 'bold', 'color': 'black'}  # Customize annotation font
)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16)

# Display the plot
plt.tight_layout()
plt.show()
```
