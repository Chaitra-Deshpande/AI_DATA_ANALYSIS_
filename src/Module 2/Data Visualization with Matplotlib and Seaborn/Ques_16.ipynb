```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 12, 8, 14, 11]
y2 = [15, 18, 10, 19, 16]
y3 = [5, 7, 3, 6, 4]

# Create a figure and axis grid (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot in the first subplot (top-left)
axes[0, 0].plot(x, y1, linestyle='-', marker='o', color='b', label='Line Plot')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X-Axis')
axes[0, 0].set_ylabel('Y1-Axis')
axes[0, 0].legend()

# Scatter plot in the second subplot (top-right)
axes[0, 1].scatter(x, y2, color='r', label='Scatter Plot')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('X-Axis')
axes[0, 1].set_ylabel('Y2-Axis')
axes[0, 1].legend()

# Bar plot in the third subplot (bottom-left)
axes[1, 0].bar(x, y3, color='g', label='Bar Plot')
axes[1, 0].set_title('Bar Plot')
axes[1, 0].set_xlabel('X-Axis')
axes[1, 0].set_ylabel('Y3-Axis')
axes[1, 0].legend()

# Empty fourth subplot (bottom-right)
axes[1, 1].axis('off')  # Turn off the axes

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the combined plot
plt.show()
```
