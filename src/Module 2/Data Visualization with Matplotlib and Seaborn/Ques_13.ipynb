```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a customized boxplot using hue and turning off the legend
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length',
    hue='species',
    palette='pastel',
    linewidth=2,
    fliersize=5,
    legend=False
)

# Add title and labels
plt.title('Distribution of Petal Lengths by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Display the plot
plt.tight_layout()
plt.show()
```
