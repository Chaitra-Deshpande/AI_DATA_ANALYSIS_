```python
# Basic Seaborn Bar Plot
# Description: Using Seaborn, plot the same bar plot from Question 1 to compare the sales data.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Product names and their corresponding sales in thousands
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [23, 75, 57, 90, 65]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the bar plot using Seaborn, with Product assigned to hue
sns.barplot(x='Product', y='Sales', hue='Product', data=df, palette='Blues', legend=False)

# Add title and labels
plt.title('Product Sales (in Thousands)')
plt.xlabel('Product')
plt.ylabel('Sales')

# Display the plot
plt.show()

```
