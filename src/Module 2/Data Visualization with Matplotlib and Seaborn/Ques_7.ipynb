```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Heights, Weights, and Sizes
data = {
    'Height': [160, 170, 165, 180, 175],
    'Weight': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]  # Represents size for each point
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the enhanced scatter plot
sns.scatterplot(
    x='Height',
    y='Weight',
    size='Size',
    sizes=(50, 300),  # Scale for size variation
    data=df,
    color='dodgerblue',
    legend=False
)

# Add title and axis labels
plt.title('Height vs Weight (Size-Enhanced)', fontsize=14)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Show the plot
plt.tight_layout()
plt.show()
```
