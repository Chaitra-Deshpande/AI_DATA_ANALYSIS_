```python
# Title: Basic Data Visualization
# Description: Plot the distribution of one feature using Seaborn.


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add species column
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Select a feature to visualize (e.g., Petal Length)
feature_to_plot = 'petal length (cm)'

# Plot distribution
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(data=iris_df, x=feature_to_plot, kde=True, bins=20, color='skyblue')
plt.title(f'Distribution of {feature_to_plot}')
plt.xlabel(feature_to_plot)
plt.ylabel('Frequency')
plt.show()
```
